```python
#Provided Code
import pandas as pd

add_df = pd.read_csv('social_ads.csv')

x_data = add_df[['Age','EstimatedSalary']].values.reshape(-1, 2)
y_target = add_df.Purchased.values
```

```python
# Put Program 4 Code Here. Do not create any additional code blocks or points may be deducted.
#CIS 343 section:01
#Program 4
#12/8/2025
#this program trains a logistic regression model and prints the coefficients and intercept in the data given as well as scores the logistic regression model in its accuracy and how many it got correct on predictions 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train, x_test, y_train, y_test = train_test_split(x_data, y_target)
log=LogisticRegression()
log.fit(X=x_train, y=y_train)
predicted=log.predict(X=x_test)
expected=y_test
print(log.coef_)
#the first coef displayed is the value that y goes up by when age is increased by one and the 2nd coef is how much an increase of estimated salary by 1 increases y
print(log.intercept_)
#this is the value of y when both age and estimated salary are zero which results in a negative number which a negative purchase doesn't really exist so it doesn't have much relevance here 
```

#### Part 2

Use pandas and Matplotlib to create a confusion matrix of the scoring of your model. Because this is a Yes/No Questionaire, you can consider it a TP/TN/FP/FN matrix.



```python
from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(y_true=expected,y_pred=predicted)
print(confusion)
print(f'{log.score(x_test,y_test):.2%}')
#based on the scoring through confusion matrix  it predicts the values correctly 84 times out of the 100 in the data set which gives the score of 84% which means the model is good at predicting the amount that will be purchased given the age and estimated salary of the participant
```
